- In the context of credit card fraud detection, outliers such as high transaction amounts or unusual patterns might actually carry critical information for identifying fraudulent activity. Removing or replacing them indiscriminately can lead to the loss of valuable insights

```python
Q1 = df['amt'].quantile(0.25)
Q3 = df['amt'].quantile(0.75)
IQR = Q3 - Q1

# Define outlier thresholds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Create a new feature 'is_outlier'
df['is_outlier'] = ((df['amt'] < lower_bound) | (df['amt'] > upper_bound)).astype(int)
```

```python

fraud_outliers = df[(df['amt'] > upper_bound) & (df['is_fraud'] == 1)]
nonfraud_outliers = df[(df['amt'] > upper_bound) & (df['is_fraud'] == 0)]

print(f"Fraudulent outliers: {len(fraud_outliers)}")
print(f"Non-fraudulent outliers: {len(nonfraud_outliers)}")
```

**StandardScaler**

```python
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
```

```python
for i in x.select_dtypes(include = 'datetime64[ns]').columns :
    x[i] = x[i].astype('int64')
```

```python
x_scaled = ss.fit_transform(x)
x_scaled
```

```python
new_x = pd.DataFrame(x_scaled,columns = x.columns)
new_x
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix,accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, r2_score
from imblearn.over_sampling import SMOTE
from sklearn.ensemble import RandomForestClassifier
```

```python
# Train-test split
X_train, X_test, y_train, y_test = train_test_split(new_x, y, test_size=0.3, random_state=42)

smote = SMOTE(sampling_strategy='auto', random_state=42)  # You can adjust 'sampling_strategy'
X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)

```

```python
model = RandomForestClassifier(n_estimators= 50)
```

```python
# Train the model
model.fit(X_train_smote, y_train_smote)
    
# Make predictions
y_pred = model.predict(X_test)
```

```python

accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

r2 = r2_score(y_test, y_pred) 
    
    
    
print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1-Score: {f1:.4f}")
print(f"RÂ² Score: {r2:.4f}")
print('-' * 60)

print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
print('-' * 60)
```
