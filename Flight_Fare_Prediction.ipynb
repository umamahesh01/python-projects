```python
from sklearn.preprocessing import LabelEncoder
```

```python

le = LabelEncoder()
```

```python

for i in df.select_dtypes(include = 'object').columns :
    df[i] = le.fit_transform(df[i])
```

```python
df
```

```python
x = df.iloc[:,:-1]
y = df['price']
x
```

```python
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()

new_x = ss.fit_transform(x)
new_x
```

```python
new_df = pd.DataFrame(new_x,columns = x.columns)
```

```python
new_df
```

```python
# days_left typically refers to the number of days between the current date (or the booking date) and the flight's departure date. It is a key feature in ticket price prediction
# because prices often vary based on how early or late a ticket is booked relative to the travel date.
```

```python
new_x = new_df[['airline', 'source_city', 'destination_city', 'stops', 'class', 'departure_time', 'days_left','duration']]
new_x 
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(new_x,y, test_size = 0.3,random_state = 42)
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
y_pred = lr.predict(X_test)
y_pred
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score


print("Mean Absolute Error (MAE):", round(mean_absolute_error(y_test, y_pred),2))
print("Mean Squared Error (MSE):", round(mean_squared_error(y_test, y_pred),2))
print("R-squared (R2):", round(r2_score(y_test, y_pred),4))

```

```python
import pickle

with open('lr_model.pkl', 'wb') as file:
    pickle.dump(lr, file)
```
